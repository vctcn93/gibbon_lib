```python
import uuid
import json
import pandas as pd
import numpy as np
from gibbon.maps import MapSensor
from gibbon.utility import Convert
```

```python
path = r'C:\Users\wenhs\Desktop\YangDaShiTouBiao\geojson\Changsha.geojson'
```

```python
with open(path, 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
features = data['features']
```

```python
buildings = list()

for f in features:
    try:
        floors = f['properties']['Floor']
        coords = f['geometry']['coordinates'][0][0][:-1]
        building = {
            'coords': coords,
            'floors': floors
        }
        buildings.append(building)
    except Exception as e:
        print(e)
```

```python
uids = [uuid.uuid4() for i in range(len(buildings))]
df = pd.DataFrame(buildings, index=uids)
```

```python
df
```

```python
f = np.vectorize(lambda a, b: np.average(np.array(a), axis=0).tolist()[b])
```

```python
df['lng'], df['lat'] = f(df['coords'], 0), f(df['coords'], 1)
```

```python
df
```

```python
origin = [112.977040, 28.208450]
radius = 2000
```

```python
sensor = MapSensor(origin, radius)
```

```python
d = df[
    (df['lng'] > sensor.llbounds[0][0]) &
    (df['lng'] < sensor.llbounds[1][0]) &
    (df['lat'] > sensor.llbounds[0][1]) &
    (df['lat'] < sensor.llbounds[1][1])
]
```

```python
def f2(lnglats):
    return [Convert.lnglat_to_mercator(ll, sensor.origin) for ll in lnglats]
```

```python
d['position'] = d['coords'].apply(f2)
d['height'] = d['floors'] * 3000
```

```python
d
```

```python
ls = list()
for i, data in d.iterrows():
    pos, heights = data['position'], data['height']
    dt = {'v': pos, 'h': heights}
    ls.append(dt)
```

```python
p =  r'C:\Users\wenhs\Desktop\YangDaShiTouBiao\geojson\buildings.json'
with open(p, 'w', encoding='utf-8') as f:
    json.dump(ls, f, ensure_ascii=False)
```
